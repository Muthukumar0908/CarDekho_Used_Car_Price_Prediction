```python
bengalore_data=pd.read_csv(r"E:\career fair\cardekho\bangalore_cars - bangalore_cars.csv.csv")
bengalore_data.head(1)
```

```python
bengalore_data['new_car_overview'][53]
```

```python
fuel_type=[]
body_type=[]
kilometer=[]
transmission_type=[]
owner_no=[]
brand=[]
model=[]
model_year=[]
price=[]
registration_year=[]
seats_count=[]
RDO_no=[]
engine_cc=[]
mileage=[]
max_power=[]
color=[]
car_engine_cc=[]
gear_box=[]
seating_capacity=[]
top_speed=[]
pickup_speed=[]
tyre_type=[]

for i in range(bengalore_data.shape[0]):
    dic_1=bengalore_data['new_car_detail'][i]
    dic_2=bengalore_data['new_car_overview'][i]
    dic_3=bengalore_data['new_car_specs'][i]
    
    res = ast.literal_eval(dic_1)
    res_1 = ast.literal_eval(dic_2)
    res_2 = ast.literal_eval(dic_3)
    
    fuel_type.append(res['ft'])
    body_type.append(res['bt'])
    kilometer.append(res['km'])
    transmission_type.append(res['transmission'])
    owner_no.append(res['ownerNo'])
    brand.append(res['oem'])
    model.append(res['model'])
    model_year.append(res['modelYear'])
    price.append(float(res['price'][2:-5])*100000)
    
    try:
        registration_year.append(res_1['top'][0]['value'][-4:])
    except:
        registration_year.append(np.nan)
    try:
        seats_count.append(res_1['top'][3]['value'][0])
    except:
        seats_count.append(np.nan)
    try:
        RDO_no.append(res_1['top'][5]['value'])
    except:
        RDO_no.append(np.nan)
    try:
        engine_cc.append(res_1['top'][7]['value'][:-3])
    except:
        engine_cc.append(np.nan)
    mileage.append(res_2['top'][0]['value'][:-5])
    try:
        max_power.append(res_2['top'][2]['value'][:-3])
    except:
        max_power.append(np.nan)
    color.append(res_2['data'][0]['list'][0]['value'])
    try:
        if (res_2['data'][0]['list'][2]['value']).isnumeric():
            car_engine_cc.append(res_2['data'][0]['list'][2]['value'])
        else:
            car_engine_cc.append(res_2['data'][0]['list'][1]['value'])
    except:
        car_engine_cc.append(np.nan)
    try:
        gear_box.append(res_2['data'][2]['list'][0]['value'][0])
    except:
        gear_box.append(np.nan)
    try:
        seating_capacity.append(res_2['data'][2]['list'][2]['value'])
    except:
        seating_capacity.append(np.nan)
    try:
        top_speed.append(res_2['data'][2]['list'][7]['value'])
    except:
        top_speed.append(np.nan)
    try:
        pickup_speed.append(res_2['data'][2]['list'][8]['value'])
    except:
        pickup_speed.append(np.nan)
    try:
        tyre_type.append(res_2['data'][2]['list'][9]['value'])
    except:
        tyre_type.append(np.nan)

bengalore_data['fuel_type']=fuel_type
bengalore_data['body_type']=body_type
bengalore_data['kilometer']=kilometer
bengalore_data['transmission_type']=transmission_type
bengalore_data['owner_no']=owner_no
bengalore_data['brand']=brand
bengalore_data['model']=model
bengalore_data['model_year']=model_year
bengalore_data['price']=price
bengalore_data['registration_year']=registration_year
bengalore_data['seats_count']=seats_count
bengalore_data['RDO_no']=RDO_no
bengalore_data['engine_cc']=engine_cc
bengalore_data['mileage']=mileage
bengalore_data['max_power(bhp)']=max_power
bengalore_data['color']=color
bengalore_data['car_engine_cc']=car_engine_cc
bengalore_data['gear_box']=gear_box
bengalore_data['seating_capacity']=seating_capacity
bengalore_data['top_speed']=top_speed
bengalore_data['pickup_speed']=pickup_speed
bengalore_data['tyre_type']=tyre_type
```

```python
fuel_type=[]
body_type=[]
kilometer=[]
transmission_type=[]
owner_no=[]
brand=[]
model=[]
model_year=[]
price=[]
registration_year=[]
seats_count=[]
RDO_no=[]
engine_cc=[]
mileage=[]
max_power=[]
color=[]
car_engine_cc=[]
gear_box=[]
seating_capacity=[]
top_speed=[]
pickup_speed=[]
tyre_type=[]

for i in range(kolkata_data.shape[0]):
    dic_1=kolkata_data['new_car_detail'][i]
    dic_2=kolkata_data['new_car_overview'][i]
    dic_3=kolkata_data['new_car_specs'][i]
    
    res = ast.literal_eval(dic_1)
    res_1 = ast.literal_eval(dic_2)
    res_2 = ast.literal_eval(dic_3)
    
    fuel_type.append(res['ft'])
    body_type.append(res['bt'])
    kilometer.append(res['km'])
    transmission_type.append(res['transmission'])
    owner_no.append(res['ownerNo'])
    brand.append(res['oem'])
    model.append(res['model'])
    model_year.append(res['modelYear'])
    price.append(float(res['price'][2:-5])*100000)
    
    try:
        registration_year.append(res_1['top'][0]['value'][-4:])
    except:
        registration_year.append(np.nan)
    try:
        seats_count.append(res_1['top'][3]['value'][0])
    except:
        seats_count.append(np.nan)
    try:
        RDO_no.append(res_1['top'][5]['value'])
    except:
        RDO_no.append(np.nan)
    try:
        engine_cc.append(res_1['top'][7]['value'][:-3])
    except:
        engine_cc.append(np.nan)
    mileage.append(res_2['top'][0]['value'][:-5])
    try:
        max_power.append(res_2['top'][2]['value'][:-3])
    except:
        max_power.append(np.nan)
    color.append(res_2['data'][0]['list'][0]['value'])
    try:
        if (res_2['data'][0]['list'][2]['value']).isnumeric():
            car_engine_cc.append(res_2['data'][0]['list'][2]['value'])
        else:
            car_engine_cc.append(res_2['data'][0]['list'][1]['value'])
    except:
        car_engine_cc.append(np.nan)
    try:
        gear_box.append(res_2['data'][2]['list'][0]['value'][0])
    except:
        gear_box.append(np.nan)
    try:
        seating_capacity.append(res_2['data'][2]['list'][2]['value'])
    except:
        seating_capacity.append(np.nan)
    try:
        top_speed.append(res_2['data'][2]['list'][7]['value'])
    except:
        top_speed.append(np.nan)
    try:
        pickup_speed.append(res_2['data'][2]['list'][8]['value'])
    except:
        pickup_speed.append(np.nan)
    try:
        tyre_type.append(res_2['data'][2]['list'][9]['value'])
    except:
        tyre_type.append(np.nan)

kolkata_data['fuel_type']=fuel_type
kolkata_data['body_type']=body_type
kolkata_data['kilometer']=kilometer
kolkata_data['transmission_type']=transmission_type
kolkata_data['owner_no']=owner_no
kolkata_data['brand']=brand
kolkata_data['model']=model
kolkata_data['model_year']=model_year
kolkata_data['price']=price
kolkata_data['registration_year']=registration_year
kolkata_data['seats_count']=seats_count
kolkata_data['RDO_no']=RDO_no
kolkata_data['engine_cc']=engine_cc
kolkata_data['mileage']=mileage
kolkata_data['max_power(bhp)']=max_power
kolkata_data['color']=color
kolkata_data['car_engine_cc']=car_engine_cc
kolkata_data['gear_box']=gear_box
kolkata_data['seating_capacity']=seating_capacity
kolkata_data['top_speed']=top_speed
kolkata_data['pickup_speed']=pickup_speed
kolkata_data['tyre_type']=tyre_type
```
